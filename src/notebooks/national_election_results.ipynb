```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load election results
df = pd.read_csv('../../data/facts/election_results_per_county.csv')

# Display as a table
display(df)

# Example: Plot total votes per party

distinct_years = df.valg_id.unique()
for year in distinct_years:
    print(f"Year: {year}")

    df_year = (df[df.valg_id == year])
    party_votes = df_year.groupby('parti_kode')['stemmer'].sum().sort_values(ascending=False)
    party_votes.plot(kind='bar', figsize=(10, 6))
    plt.xlabel('Party')
    plt.ylabel('Total Votes')
    plt.title('Total Votes per Party')
    plt.tight_layout()
    plt.show()
```
